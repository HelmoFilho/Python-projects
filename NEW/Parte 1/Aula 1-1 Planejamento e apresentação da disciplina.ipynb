## 1.   *Importações dos módulos gerais*

    * Aqui será importado o numpy. Ele será utilizado tanto para criação como multiplicação de matrizes.




```
import numpy as np
```

## 2.  Criação e multiplicação de matrizes


    * Criação de Matrizes

```
#Será criado uma matriz 3x3. Os valores da mesma serão aleatórios para cada execução do código porêm os valores 
#serão variados para cada execução da célula.
Matriz_1 = np.random.random((3, 3))
Matriz_1
```

```
#A matriz 2 será criada com o mesmo método da Matriz_1.
Matriz_2 = np.random.random((3, 3))
Matriz_2
```

    * Multiplicação de Matrizes

```
Mult_M = np.dot(Matriz_1, Matriz_2)
Mult_M
```
