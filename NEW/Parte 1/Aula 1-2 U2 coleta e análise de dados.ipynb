*   Descrever no caso dos dados da Iris qual dos 4 atributos é melhor para a diferenciação das classes.

## 1.   *Importações dos módulos gerais*

    * Aqui serão importados o numpy, pandas, matplotlib e o seaborn




```
# This is not strictly for a Python rather. 
%matplotlib inline
```

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
 
sns.set()                         #Configura os plots do matplotlib para terem os mesmos estilos do seaborn
sns.set_style("whitegrid")        #Alterando o estilo do plot para melhor vizualização no colab (Tema escuro fica ruim de ver o fundo)
```

## 2.   *Dataset "Iris"*

      *   O dataset será retirado de um repositório pessoal do GITHUB


```
Iris = pd.read_csv('https://raw.githubusercontent.com/Berawecker/Datasets/main/Iris.csv')
Iris = Iris.drop(["Id"], axis = 1)
 
Iris
```

#### 3.   *Histograma do dataset "Iris"*

    * Sem diferenciação por classe nos histogramas

```
fig, ax = plt.subplots(2, 2, figsize = (10,10))
sns.histplot(data = Iris, x = "SepalLengthCm", bins = 10, ax = ax[0][0])
sns.histplot(data = Iris, x = "SepalWidthCm" , bins = 10, ax = ax[0][1])
sns.histplot(data = Iris, x = "PetalLengthCm", bins = 10, ax = ax[1][0])
sns.histplot(data = Iris, x = "PetalWidthCm" , bins = 10, ax = ax[1][1])
```

Analisando os histogramas dos 4 atributos do dataset IRIS, percebe-se que:

  * O atributo "SepalLengthCm" possui seus dados muito agrupados. Não se pode inferir com certeza quais dados pertecem a quais classes ou mesmos quantas classes existem por este gráfico (em uma situação em que não se sabe quantas classes têm).

  * O atributo "SepalWidthCm", sofre do mesmo problema que o atributo anterior: Dados muito agrupados.

  * O atributo "PetalLengthCm", Parece ser o melhor para separar os dados visto que um pequeno grupo de dados está separado do resto. É possivel deduzir que este grupo pertence a uma classe e o resto das classes estão no outro grupo.

  * O atributo "PetalWidthCm", tambêm poderia se tornar um bom candidato para classificação univariável, porém, diferente do "PetalLengthCm", é possível perceber que não existem grupos totalmente isolados neste caso. 

Então, analisando os histogramas acima é possivel dizer que o atributo "PetalLengthCm" é o mais adequado para a diferenciação de classes.

    * Com diferenciação por classe nos histogramas

```
#Serão utilizados multiplos plots do histograma para não ter de realizar um pairplot e para ficar mais visível
fig, ax = plt.subplots(2, 2, figsize = (10,10))
sns.histplot(data = Iris, x = "SepalLengthCm", bins = 10, hue= "Species", ax = ax[0][0])
sns.histplot(data = Iris, x = "SepalWidthCm" , bins = 10, hue= "Species", ax = ax[0][1])
sns.histplot(data = Iris, x = "PetalLengthCm", bins = 10, hue= "Species", ax = ax[1][0])
sns.histplot(data = Iris, x = "PetalWidthCm" , bins = 10, hue= "Species", ax = ax[1][1])
```

Analisando os histogramas dos 4 atributos do dataset IRIS, percebe-se que:

  * O atributo "SepalLengthCm" Possui muitos dos dados sobrepostos. Seria muito complicado definir as classes somente com este atributo.

  * O atributo "SepalWidthCm" sofre do mesmo problema do atributo anterior: Muitos dados de uma determinada classe sobrepostos.

  * O atributo "PetalLengthCm" pode ser usado para separa totalmente a classe "Iris-setosa" das outras duas, porêm existe uma sobreposição de mais de 30 dados das classes "Iris-versicolor" e "Iris-virginica".

  * O atributo "PetalWidthCm" pode, **aparentemente** (visto que não há sobreosição com outras classes), ser utilizado para separar a classe "Iris-setosa" das outras duas. Existe uma sobreposição de dados entre as outras duas classes, mas ela aparenta ser de menor volume que a do atributo "PetalLengthCm".

Então, analisando os histogramas acima, pode-se dizer que:

* Se a intenção for uma classificação total da classe "Iris-setosa", pode-se utilizar o atributo "PetalLengthCm"

* Se a intenção for uma classificação que pode ter alguma sobreposição em todas as classes, mas de muita menor intensidade nas classes "Iris-versicolor" e "Iris-virginica" e claro uma sobreposição quase nula na classe "Iris-setosa", pode-se escolher o atributo "PetalWidthCm"

Como mencionado acima, o atributo "PetalWidthCm" se torna melhor na classificação de todas as classes visto que esta têm uma sobreposição muito menos dos dados das classes "Iris-versicolor" e "Iris-virginica". Por isto, esta seria a melhor no caso de uma classificação univariável.
