#### 5. K-Means Clustering

    * Separação dos dados

```
f_iris = iris.iloc[:, :4].values
t_iris = iris.iloc[:, 4].values
```

```
f_wine = wine.iloc[:, 1:]
t_wine = wine.iloc[:, 0].values
```

Os dados de cada um dos datasets foram separados em seus respectivos "targets" e "features", apesar dos targets não serem utilizados devido a natureza "não-supervisionada" do algoritmo. 

    * Kmeans e visualização dos cluters com os dados da IRIS

Será utilizado o método do "cotovelo" ou "elbow" (devido ao seu formato) para definir a melhor quantidade de clusters. O método do "elbow" consiste em somar o erro quadrático de todos os clusters com os pontos e conforme o numero de cluster tende a aumentar, o somatório do erro diminui também. A partir deste gráfico se escolhe qual o melhor valor para realizar a clusterização.

```
wcss_1 = []   #within cluster sum of squares

for i in range(1, 11):
  kmeans_iris = KMeans(n_clusters = i)
  kmeans_iris.fit(f_iris)
  wcss_1.append(kmeans_iris.inertia_)
```

```
plt.plot(range(1,11), wcss_1, marker = "o")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method - Iris Dataset")
```

Observando o gráfico acima, se percebe que o "k" com valor de 2 já possue uma grande diferença do k = 1 (pior caso). mas k = 3 está mais proximo da zona em que não há uma variação tão grande entre os resultados então o "k" escolhido será 3

```
kmeans_iris = KMeans(n_clusters = 3)
kmeans_iris_pred = kmeans_iris.fit_predict(f_iris)
```

```
f_iris[kmeans_iris_pred == 0, 0]
```

```
plt.scatter(f_iris[kmeans_iris_pred == 0, 0], f_iris[kmeans_iris_pred == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(f_iris[kmeans_iris_pred == 1, 0], f_iris[kmeans_iris_pred == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(f_iris[kmeans_iris_pred == 2, 0], f_iris[kmeans_iris_pred == 2, 1], s = 100, c = 'green', label = 'Cluster 3')

plt.scatter(kmeans_iris.cluster_centers_[:, 0], kmeans_iris.cluster_centers_[:,1], s = 100, c = 'yellow', label = 'Centroids')

plt.legend()
```

    * Kmeans com os dados da WINE

```
wcss_2 = []

for i in range(1, 11):
  kmeans_wine = KMeans(n_clusters = i)
  kmeans_wine.fit(f_wine)
  wcss_2.append(kmeans_wine.inertia_)
```

```
plt.plot(range(1,11), wcss_2, marker = "o")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method - Wine Dataset")
```

```
kmeans_wine = KMeans(n_clusters = 3)
kmeans_wine_pred = kmeans_wine.fit_predict(f_wine)
```
