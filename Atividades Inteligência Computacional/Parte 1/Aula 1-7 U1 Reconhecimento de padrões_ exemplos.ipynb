#### 4. Encontrando o melhor valor de vizinhos (n_neighbors)

    * Será a separação dos previsores e das classes

```
x_iris = iris.iloc[:,:4].values
y_iris = iris.iloc[:,4].values

x_diabetes = diabetes.iloc[:, :8].values
y_diabetes = diabetes.iloc[:, 8].values
```

    * Encontrando o melhor valor de "n_neighbors"

```
#Escolhendo o valor máximo de n_neighbors(k)
k_max_iris = int(round(np.sqrt(len(y_iris))))   #O "k" é geralmente escolhido pela raiz quadrada da quantidades de dados usados para o treino
k_max_diabetes = int(round(np.sqrt(len(y_diabetes))))

if k_max_iris % 2 == 0:                          #É importante também que ele seja impar para não acontecer um empate na classificação
  k_max_iris += 1

if k_max_diabetes % 2 == 0:                          
  k_max_diabetes += 1

print(f"k máximo iris = {k_max_iris}")
print(f"k máximo diabetes = {k_max_diabetes}")
```

```
def KNN_k_find(x , y, k_maximo):
  
  accuracy = []

  for _ in range(1, k_maximo + 1, 2):

    clf_knn = KNN(n_neighbors = _)
    clf_knn.fit(x, y)
    test_predict = clf_knn.predict(x)
    accuracy.append(accuracy_score(y, test_predict) * 100)

  return accuracy
```

```
#Plotando o gráfico de acurácia do iris
iris_k_list = KNN_k_find(x_iris, y_iris, k_max_iris)

plt.figure(figsize=(12, 6))
plt.plot(range(1, k_max_iris + 1, 2), iris_k_list, color = 'red', linestyle = 'dashed', marker = 'o', markerfacecolor = 'blue', markersize = 5)
plt.title('Accuracy of different K values for Iris dataset')
plt.xlabel('K Value')
plt.ylabel('Accuracy')
```

O melhor valor de do parâmetro "n_neighbors" é 9. Ele será utilizado para criar o classificador.

```
#Plotando o gráfico de acurácia do diabetes
iris_k_list = KNN_k_find(x_diabetes, y_diabetes, k_max_diabetes)

plt.figure(figsize=(12, 6))
plt.plot(range(1, k_max_diabetes + 1, 2), iris_k_list, color = 'red', linestyle = 'dashed', marker = 'o', markerfacecolor = 'blue', markersize = 5)
plt.title('Accuracy of different K values for Diabetes dataset')
plt.xlabel('K Value')
plt.ylabel('Accuracy')
```

O melhor valor de do parâmetro "n_neighbors" é 3. Ele será utilizado para criar o classificador.

#### 5. Classificador KNN com os dados da iris

```
#Treinando o classificador
clf_knn = KNN(n_neighbors = 9)
clf_knn.fit(x_iris, y_iris)

iris_predict = clf_knn.predict(x_iris)
```

```
#Plotando a matriz de confusão
iris_matrix = confusion_matrix(y_iris, y_predict)
heatmap_iris = pd.DataFrame(iris_matrix, columns=np.unique(y_iris), index = np.unique(y_iris))

heatmap_iris.index.name = 'Real'
heatmap_iris.columns.name = 'Previsto'
plt.title("Matriz de Confusão do iris", fontsize = 15)

sns.heatmap(heatmap_iris, annot=True)
```

```
#Mostrando o accuracy_score do modelo
print("Pontuação de precisão do iris: {}%".format(accuracy_score(y_iris ,y_predict)*100))
```

#### 6. Classificador KNN com os dados do Diabetes

```
#Treinando o classificador
clf_knn_diabetes = KNN(n_neighbors = 3)
clf_knn_diabetes.fit(x_diabetes, y_diabetes)

diabetes_predict = clf_knn_diabetes.predict(x_diabetes)
```

```
#Plotando a matriz de confusão
diabetes_matrix = confusion_matrix(y_diabetes, diabetes_predict)
diabetes_matrix = pd.DataFrame(diabetes_matrix, columns=np.unique(y_diabetes), index = np.unique(y_diabetes))

diabetes_matrix.index.name = 'Real'
diabetes_matrix.columns.name = 'Previsto'
plt.title("Matriz de Confusão do diabetes", fontsize = 15)

sns.heatmap(diabetes_matrix, annot = True)
```

```
#Mostrando o accuracy_score do modelo
print("Pontuação de precisão do Diabetes: {}%".format(accuracy_score(y_diabetes, diabetes_predict)*100))
```
