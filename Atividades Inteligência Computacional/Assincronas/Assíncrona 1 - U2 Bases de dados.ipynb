## 1.   *Importações dos módulos*

    * Aqui serão importados o numpy, pandas, matplotlib e o seaborn




```
# This is not strictly for a Python rather. 
%matplotlib inline
```

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
 
sns.set()                         #Configura os plots do matplotlib para terem os mesmos estilos do seaborn
sns.set_style("whitegrid")        #Alterando o estilo do plot para melhor vizualização no colab (Tema escuro fica ruim de ver o fundo)
```

## 2.   *Dataset "Iris"*

      *   O dataset será retirado da biblioteca sklearn


```
iris = load_iris()
x = iris['data']
y = iris['target']

iris = pd.DataFrame(data = x, columns = iris['feature_names'])
iris["species"] = y
iris["species"] = iris["species"].replace(to_replace= [0, 1, 2], value = ['setosa', 'versicolor', 'virginica'])
iris
```

#### 3.   *Pairplot do dataset "Iris"*

```
#Plotando os dados em um "pairplot"
sns.pairplot(iris, hue = "species", palette = "brg", markers = ["o", "s", "D"], diag_kind = "hist")
```
